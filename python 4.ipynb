```python
class Movie:
    def __init__(self, title, year):
        self.title = title
        self.year = year

    def __str__(self):
        return f" {self.title} ({self.year})"

    def __repr__(self):
        return f" Movie ('{self.title}', {self.year})"

m = Movie("Inception", 2010)

print(m)           # Output: Inception (2010)
print(repr(m))     # Output: Movie('Inception', 2010)
```

```python
class Account:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance
```

```python
class SavingsAccount(Account):
    def add_interest(self):
        self.deposit(self.get_balance() * 0.03)
class CurrentAccount(Account):
    def add_interest(self):
        self.deposit(0)
```

```python
def credit_interest(account):
    account.add_interest()

s = SavingsAccount("Tanya", 1000)
c = CurrentAccount("Tanya", 1000)

credit_interest(c)
credit_interest(s)

print("Savings balance:", s.get_balance())  
print("Current balance:", c.get_balance()) 
```

```python
class Mother:
    def __init__(self, mother_name):
        self.mother_name = mother_name

class Father:
    def __init__(self, father_name):
        self.father_name= father_name

class Child(Father, Mother):
    def __init__(self, mother_name, father_name, child_name):
        Mother.__init__(self, mother_name)
        Father.__init__(self, father_name)
        self.child_name = child_name

child = Child("Eve", "Adam", "cain")
print(child.mother_name, child.father_name, child.child_name)
```
